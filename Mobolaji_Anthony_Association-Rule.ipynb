```python
# Normalization cheking for null values 
Online.sort_values(['StockCode','Description'], ascending=False).groupby('Quantity').head()
```

```python
# Cleaning and removing null values 
Online['Description'] = Online['Description'].str.strip()
Online.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
Online['InvoiceNo'] = Online['InvoiceNo'].astype('str')
Online = Online[~Online['InvoiceNo'].str.contains('C')]
Online
```

```python
# Defining the quantity 
Cart = (Online[Online['Country'] == 'France']
       .groupby(['InvoiceNo', 'Description'])['Quantity']
       .sum().unstack().reset_index().fillna(0)
       .set_index('InvoiceNo'))
Cart
```

```python
# encoding and seting all paramiters right removing anomalies 
def encode_units(x):
    if x <1:
      return 0
    else:
      return 1
Cart_set = Cart.applymap(encode_units)
Cart_set 
```

```python
cartdf = pd.get_dummies(Cart_set)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# getting the most frequent item set 
Frequency = apriori (cartdf, min_support=0.02, use_colnames= True)
lift = association_rules(Frequency, metric = 'lift', min_threshold = 1)
lift.head()
```

```python
rules = association_rules(Frequency,metric = 'lift' , min_threshold=2)
rules[(rules['lift'] >= 2)& (rules['confidence'] > 0.1)]
```
